```python
import pandas as pd
import json

# Read the JSON file
with open('snopes_fact_checks.json', 'r') as file:
    data = json.load(file)

# Convert the JSON data to a pandas DataFrame
df = pd.DataFrame(data)

# Clean the ratings
def clean_rating(rating):
    rating = rating.split('\n')[0].strip()  # Remove 'About this rating' and any whitespace
    if rating in ['True', 'False', 'Mixture', 'Correct Attribution']:
        return rating
    elif rating == 'Originated as Satire':
        return 'False'
    elif rating in ['Fake', 'Unfounded', 'Labeled Satire']:
        return 'False'
    elif rating == 'Unproven':
        return 'Mixture'
    else:
        return 'Mixture'  # Default case for any unexpected ratings

# Apply the cleaning function to the 'Rating' column
df['Rating'] = df['Rating'].apply(clean_rating)

df.head()
```

```python
import pandas as pd
from transformers import pipeline
import torch

class DifficultyCategorizer:
    def __init__(self, model_name="facebook/bart-large-mnli", device=-1):
        """
        Initializes the zero-shot classification pipeline.

        Args:
            model_name (str): Hugging Face model name for zero-shot classification.
            device (int): Device to run the model on (-1 for CPU, >=0 for GPU).
        """
        self.classifier = pipeline("zero-shot-classification", model=model_name, device=device)

    def categorize(self, text, candidate_labels):
        """
        Categorizes the input text into one of the candidate labels.

        Args:
            text (str): The text to classify.
            candidate_labels (list): List of labels to classify the text into.

        Returns:
            str: The label with the highest confidence score.
        """
        result = self.classifier(text, candidate_labels)
        return result['labels'][0]

def main():
    # Load the cleaned JSON data into a pandas DataFrame
    df = pd.read_json('snopes_fact_checks.json', orient='records')

    # Initialize the DifficultyCategorizer
    # Set device=0 if you have a GPU available, else leave as default (CPU)
    categorizer = DifficultyCategorizer(model_name="facebook/bart-large-mnli", device=-1)

    # Define candidate labels
    candidate_labels = ["Easy", "Hard to Know"]

    # Apply categorization to the 'Summary' column
    print("Categorizing difficulty levels. This may take a while...")
    df['Difficulty'] = df['Summary'].apply(lambda x: categorizer.categorize(x, candidate_labels))

    # Display the first few rows to verify
    print(df[['Title', 'Summary', 'Difficulty']].head())

    # Save the updated DataFrame to a new CSV file
    df.to_csv('backend/snopes_fact_checks_with_difficulty.csv', index=False)
    print("Categorization complete. Saved to 'snopes_fact_checks_with_difficulty.csv'.")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import json

# Read the JSON file
with open('snopes_fact_checks.json', 'r') as file:
    data = json.load(file)

# Convert the JSON data to a pandas DataFrame
df = pd.DataFrame(data)

# Clean the ratings
def clean_rating(rating):
    if pd.isna(rating):
        return None
    rating = rating.split('\n')[0].strip()  # Remove 'About this rating' and any whitespace
    return rating if rating != 'N/A' else None

# Apply the cleaning function to the 'Rating' column
df['Rating'] = df['Rating'].apply(clean_rating)

# Drop rows where Rating is None (this includes original N/A and empty values)
df = df.dropna(subset=['Rating'])

# Display the first few rows to verify
print(df[['Title', 'Rating']].head())
df_sorted = df.sort_values(by='Date', ascending=False)

# Save the updated DataFrame to a new CSV file
df_sorted.to_csv('snopes_fact_checks_cleaned.csv', index=False)
print("Cleaning complete. Saved to 'snopes_fact_checks_cleaned.csv'.")

df_sorted.to_json('snopes_fact_checks_cleaned.json', orient='records')
print("JSON file saved as 'snopes_fact_checks_cleaned.json'.")
```
